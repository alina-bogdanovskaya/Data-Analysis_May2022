```python
%matplotlib inline
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema
import math
```

Реализовать оптимизацию методом градиентного спуска для выражения $x^2(np.sin(0,5_x)^2+1)$. Какие параметры шага и количества итераций оптимальны? При каких значениях шага оптимизиция не выходит из локальных минимумов?

```python
def func2(_x):
  return (_x**2) * (np.sin(0.5 * _x)**2+1)
  

def func_deriv(_x):
  return _x * (2 * np.sin(0.5*_x)**2 + 0.5 * _x * np.sin(_x) + 2)
```

```python
def draw_plot(x_val, y_val):
  x = np.linspace(-50, 50, 5000)
  y = func2(x)

  arr_x = np.array(x_val)
  arr_y = np.array(y_val)

  d = argrelextrema(y, np.less)

  f = plt.figure()
  f.set_figwidth(20)
  f.set_figheight(10)

  plt.plot(x, y)
  plt.plot(x[d], y[d], "x", label="local min", c='g')
  plt.vlines(x[d], -200, 4500, colors='k', linestyles='dashed', alpha=0.5)
  plt.scatter(arr_x, arr_y, c='orange')
  plt.legend()
  
```

```python
x0_list = [np.random.uniform(-50, 50) for i in range(100000)]
```

```python
alpha = 0.06
n_iter = 50

converg_1 = []
diverg_1= []
stuck_1 = [] 

results_1 = {}

for el in x0_list:
  x = el
  results_1[el] = []
   
  for i in range(n_iter):
    results_1[el].append((i, x, func2(x), func_deriv(x)))
    x -= alpha*func_deriv(x)

  if abs(x) > 50 or math.isnan(x) or math.isfinite(x) == False:
    diverg_1.append(el)

  elif math.isclose(0, x, abs_tol=1):
    converg_1.append(el) 

  else:
    stuck_1.append(el)

print(f'convergence: {len(converg_1)}, divergence: {len(diverg_1)}, stucks: {len(stuck_1)}')
```

При $\alpha$ 0.06 примерно в 17% случаев метод расходится, застревания в локальных минимумах не наблюдается. Для поиска глобального минимума достаточно сравнительно небольшого количества итераций - 50.

```python
alpha = 0.05
n_iter = 100

converg_2 = []
diverg_2 = []
stuck_2 = [] 

results_2 = {}

for el in x0_list:
  x = el
  results_2[el] = []
   
  for i in range(n_iter):
    results_2[el].append((i, x, func2(x), func_deriv(x)))
    x -= alpha*func_deriv(x)

  if abs(x) > 50 or math.isnan(x) or math.isfinite(x) == False:
    diverg_2.append(el)

  elif math.isclose(0, x, abs_tol=1):
    converg_2.append(el) 

  else:
    stuck_2.append(el)

print(f'convergence: {len(converg_2)}, divergence: {len(diverg_2)}, stucks: {len(stuck_2)}')
```

При $\alpha$ 0.05 примерно в 10-11% случаев метод расходится, застревания в локальных минимумах не наблюдается. При данном значении $\alpha$ оптимальное количество итераций 100.

```python
x0 = converg_2[random.randint(0, len(converg_2))]

X_2, Y_2 = [], []

for item in results_2[x0]:
  X_2.append(item[1]), 
  Y_2.append(item[2])


draw_plot(X_2, Y_2)
```
